```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_parquet("final_dataset.parquet")
total_df = df
```

```python
total_df['length'] = total_df['article'].apply(lambda x: len(x.split()))
```

```python
# Set a seaborn style
sns.set_style("whitegrid")

# Create a larger figure
plt.figure(figsize=(10, 6))

# Plot histogram with KDE
sns.histplot(total_df['length'], bins=50, kde=True, color='blue', edgecolor='black', alpha=0.7)

# Labels and title
plt.xlabel("Length of Articles", fontsize=14, fontweight='bold')
plt.ylabel("Frequency", fontsize=14, fontweight='bold')
plt.title("Distribution of Article Lengths", fontsize=16, fontweight='bold')

# Show the plot
plt.show()
```

```python
total_df.drop(columns=['length'], inplace=True)
```

```python
from collections import defaultdict
from tqdm import tqdm
import notification_nltf as ntf
from wordcloud import WordCloud
from nltk.corpus import stopwords
import nltk
from string import punctuation
import re
nltk.download('stopwords')
```

```python
corpus_size = set()
corpus_without_stopwords = set()
word_count = defaultdict(int)

stop_words = set(stopwords.words('english'))

def clean_word(word):
    return re.sub(rf"[{punctuation}]", "", word).lower()

# Iterate through articles
for _, row in tqdm(total_df.iterrows(), total=len(total_df)):  
    article = row['article'].split()
    for word in article:
        cleaned_word = clean_word(word) 
        if cleaned_word:
            corpus_size.add(cleaned_word)
            if cleaned_word not in stop_words:
                corpus_without_stopwords.add(cleaned_word)
                word_count[cleaned_word] += 1

# Send notification when processing is done
ntf.send_notification(message="Word count completed.", topic='vscode_tester')
```

```python
wordcloud = WordCloud(width=800, height=400, background_color="white", colormap="viridis").generate_from_frequencies(word_count)


plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud of Article Text", fontsize=14, fontweight="bold")
plt.show()
```

```python
print(f"Corpus size without removing stop words : {len(corpus_size)}")
print(f"Corpus size after removing stop words : {len(corpus_without_stopwords)}")
print(f"The Percentage of stop words in the corpus: {(1 - len(corpus_without_stopwords) / len(corpus_size))*100}%")
```
